## <미션1> 폴더내의 자료 확인
파이썬 glob 명령어에 과 파이썬 os.listdir 의 차이점 정리</br>
glob 또는 os.listdir 명령어를 이용하여 'flower.jpg' 파일의 작업 디렉토리에 있으면 '자료있음', 없으면' 자료없음'으로 출력

1. glob 모듈은 파일명을 리스트 형식으로 반환 한다. 다만 정규식 사용이 어렵고 엑셀 등에서도 사용할 수 있는 '*'와 '?'같은 와일드 카드만 지원한다.
2. os.listdir 매써드는 지정한 디렉토리 내의 모든 파일과 디렉토리를 리스트(list)로 리턴한다.

차이점이라 하면, glob은 지정한 파일명을 리스트로 리턴하고 정규식 사용이 불가한점, os.listdir는 디렉토리의 모든 파일을 리스트로 반환하는점 이라고 볼 수 있다.

```python
import os

path = os.getcwd()
dir_list = os.listdir(path)

for i in dir_list:
    if i == 'flower.jpg':
        print('flower.jpg 자료있음')
        

```

## <미션4> 이미지 배열값 조정
tmp1=imgArr 과 tmp1=imgArr.copy() 의 차이점에 대하여 알기 # 구글검색, 파이썬 깊은복사</br>
tmp1이미지 -> 이미지의 모든 배열숫자값이 200이상이면 0으로 변경함</br>
tmp2이미지 -> 이미지의 모든 배열숫자값이 100이하이면 255로 변경함</br>

```python
tmp1=imgArr.copy()   # 
print(tmp1[0,:5])

# 미션:   tmp1이미지 -> 이미지의 모든 배열숫자값이 200이상이면 0으로 변경함
tmp1[tmp1>=200] =0
a = tmp1

plt.imshow(a)
```
